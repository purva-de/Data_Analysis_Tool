```python
from bs4 import BeautifulSoup
import requests

# URL to scrape
url = 'https://www.woodenstreet.com/floor-lamps'

# Send request with user agent to avoid blocking
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})

# Check if request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find all product blocks based on the 'load-product-item' class
    products = soup.find_all('div', class_='load-product-item')
    
    # Check if any products are found
    if not products:
        print("No products found! The content might be dynamically loaded.")
    
    # Loop through each product
    for product in products:
        try:
            # Extract the product name from the 'heading' class inside an <h3> tag
            product_name_tag = product.find('h3', class_='heading')
            product_name = product_name_tag.text.strip() if product_name_tag else "Product name not found"
            
            # Extract the product price from the <small> tag inside the 'price' class
            price_tag = product.find('small')
            product_price = price_tag.text.strip() if price_tag else "Product price not found"
            
            # Print the product name and price
            print(f"Product Name: {product_name}")
            print(f"Product Price: ₹{product_price}")
            print('-' * 40)
        
        except Exception as e:
            print(f"Error extracting data: {e}")
else:
    print(f"Failed to retrieve page. Status code: {response.status_code}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up Selenium WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# URL to scrape
url = 'https://www.woodenstreet.com/floor-lamps'

# Open the URL
driver.get(url)

# Wait for the page to load
time.sleep(5)

# Extract all products and their prices
products = driver.find_elements(By.CLASS_NAME, 'load-product-item')

for product in products:
    try:
        product_name = product.find_element(By.CLASS_NAME, 'heading').text.strip()
    except:
        product_name = "Product name not found"
    
    try:
        product_price = product.find_element(By.TAG_NAME, 'small').text.strip()
    except:
        product_price = "Product price not found"
    
    print(f"Product Name: {product_name}")
    print(f"Product Price: ₹{product_price}")
    print('-' * 40)

# Close the browser window
driver.quit()
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.woodenstreet.com/floor-lamps'

# Send request to the URL
response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})

# Check if request was successful
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Loop through multiple pages (assuming there are pagination links)
    for page in range(1, 6):  # Adjust range based on the number of pages
        page_url = f'{url}?page={page}'  # Example URL for pagination
        page_response = requests.get(page_url, headers={'User-Agent': 'Mozilla/5.0'})
        
        if page_response.status_code == 200:
            page_soup = BeautifulSoup(page_response.text, 'html.parser')
            products = page_soup.find_all('div', class_='load-product-item')

            for product in products:
                product_name = product.find('h3', class_='heading')
                product_price = product.find('small')

                if product_name and product_price:
                    print(f"Product Name: {product_name.text.strip()}")
                    print(f"Product Price: {product_price.text.strip()}")
                    print('-' * 40)
        else:
            print(f"Failed to retrieve page {page}. Status code: {page_response.status_code}")
else:
    print(f"Failed to retrieve the main page. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the website
url = "https://nestasia.in/collections/home-decor"

# Send a GET request
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find all product elements
products = soup.find_all('div', class_='result-product-item-info')

# If products are not found, print some HTML for debugging
if not products:
    print("No products found. HTML content for inspection:")
    print(soup.prettify())  # Print the parsed HTML

# Create a list to store the data
product_data = []

# Iterate through each product and extract name and price
for product in products:
    # Extract product name
    name = product.find('h3', class_='product-item-title')
    price = product.find('span', class_='wizzy-pro-price')

    # Handle cases where name or price is missing
    if name:
        name = name.text.strip()
    else:
        name = "N/A"

    if price:
        price = price.text.strip()
    else:
        price = "N/A"

    # Store in list as a dictionary
    product_data.append({'Product Name': name, 'Price': price})

# Display in a structured format
print("{:<50} {:<10}".format('Product Name', 'Price'))
print('-' * 60)
for product in product_data:
    print("{:<50} {:<10}".format(product['Product Name'], product['Price']))
```

```python
def price_category(price):
    if price < 1000:
        return 'Low Price'
    elif 1000 <= price < 3000:
        return 'Medium Price'
    else:
        return 'High Price'

# Apply the function to create a new 'Category' column
df['Category'] = df['Price'].apply(price_category)

# Display the data with categories
print(df.head())
```

```python
# Sample categorization (you can expand based on product names)
def categorize_product(name):
    if 'Candle' in name.lower():
        return 'Candle'
    elif 'Bowls' in name.lower():
        return 'Bowls'
    elif 'Statue' in name.lower():
        return 'Statue'
    elif 'Mirror' in name.lower():
        return 'Mirror'
    elif 'Showpiece' in name.lower():
        return 'Showpiece'
    elif 'Glass Platter' in name.lower():
        return 'Glass Platter' 
    elif 'Tealight' in name.lower():
        return 'Tealight'     
    else:
        return 'Other'

df['Category'] = df['Product Name'].apply(categorize_product)

# Check the categorization
print(df[['Product Name', 'Category']].head())
```
